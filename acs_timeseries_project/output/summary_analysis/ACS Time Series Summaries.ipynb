# ACS TIME SERIES BROAD SUMMARIES

```python
import os, csv, pandas as pd
```

```python
sumlev_names={'010':'US','020':'Region','030':'Division','040':'State','050':'County',
             '060':'County Subdiv','140':'Tract','150':'Blk Group','160':'Place','170':'Cons City',
              '250':'Tribal Area','251':'Tribal Subdiv','256':'Tribal Tract','258':'Tribal Blk Group', 
              '310':'CBSA','400':'Urban Area','500':'Cong Dist','795':'PUMA','860':'ZCTA'}
```

```python
tab_names={}
tab_file=os.path.join('metatables','acs_tables.csv')
with open(tab_file, 'r', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        tab_names[row['tab_id'].lower()]=row['tab_lbl'].title()
        
droptabs=['b09001','b09020','b98001','b98003']
for k in [k for k in tab_names if k in droptabs]: del tab_names[k]
len(tab_names)
```

```python
var_names={}
var_file=os.path.join('metatables','acs_variables.csv')
with open(var_file, 'r', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        var_names[row['var_id'].lower()]=row['var_lbl'].title()
        
dropvars=['b09001_001','b09020_001','b98001_002','b98003_001']
for k in [k for k in var_names if k in dropvars]: del var_names[k]

var_names_short={}
for k,v in var_names.items():
    short_label=v.split('%')[-2:]
    var_names_short[k]=' - '.join(short_label)    
len(var_names)
```

## CV by Summary Level

```python
cv_pctgeo_file=os.path.join('sumdata','change_cv_pct_geo.tsv')
cv_pctgeo=pd.read_csv(cv_pctgeo_file,delimiter='\t',dtype={'sumlevel':str})
cv_pctgeo['sumlevel']=cv_pctgeo['sumlevel'].map(sumlev_names)
cv_pctgeo.loc[cv_pctgeo['change_cv'].isin(['medium','high'])].groupby(by='sumlevel').sum().sort_values('percent',ascending=False)
```

## CV by Table

```python
cv_pcttab_file=os.path.join('sumdata','change_cv_pct_tab.tsv')
cv_pcttab=pd.read_csv(cv_pcttab_file,delimiter='\t',dtype={'sumlevel':str})
cv_pcttab['label']=cv_pcttab['table'].map(tab_names)
cv_pcttab.loc[cv_pcttab['change_cv'].isin(['medium','high'])].groupby(by='table').sum().sort_values('percent',ascending=False)
```

## CV by Type of Variable

```python
cv_pctvtype_file=os.path.join('sumdata','change_cv_pct_vtype.tsv')
cv_pctvtype=pd.read_csv(cv_pctvtype_file,delimiter='\t')
cv_pctvtype
```

## CV by Variable

```python
cv_pctvar_file=os.path.join('sumdata','change_cv_pct_var.tsv')
cv_pctvar=pd.read_csv(cv_pctvar_file,delimiter='\t')
cv_pctvar['label']=cv_pctvar['variable'].map(var_names_short)
cv_pctvar.loc[cv_pctvar['change_cv'].isin(['medium','high'])].groupby(
    by='variable').sum().sort_values('percent',ascending=False).head(n=20)
```

```python
cv_pctvar.loc[cv_pctvar['change_cv'].isin(['medium','high'])].groupby(by='variable').sum().sort_values(
    ['percent','change_cv_count'],ascending=False).tail(n=20)
```

```python
grouped_df=cv_pctvar.loc[cv_pctvar['change_cv'].isin(['medium','high'])].groupby(by='variable').sum()
grouped_df.loc[grouped_df['percent']>=20.0].count()
```

## CV by Table and Summary Level

```python
gfilter='050'
print(sumlev_names[gfilter].upper())
cv_pcttabgeo_file=os.path.join('sumdata','change_cv_pct_tabgeo.tsv')
cv_pcttabgeo=pd.read_csv(cv_pcttabgeo_file,delimiter='\t',dtype={'sumlevel':str})
cv_pcttabgeo['label']=cv_pcttabgeo['table'].map(tab_names)
cv_pcttabgeo.loc[(cv_pcttabgeo['change_cv'].isin(['medium','high'])) 
                 & (cv_pcttabgeo['sumlevel']==gfilter)].groupby(by=['table','sumlevel']).sum().sort_values('percent',ascending=False)
```
